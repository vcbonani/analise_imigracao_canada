```python
import pandas as pd

url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv'

dados = pd.read_csv(url)
```

```python
dados
```

```python
dados.info()
```

## Analisar as tendências de imigração do Brasil ao longo dos anos

```python
#trocando o índice da numeração pela coluna País
dados.set_index('País', inplace = True)
dados
```

```python
#criando variável do tipo lista com intervalo de anos usando função range e map
anos = list(map(str, range(1980, 2014)))
anos
```

```python
#buscando somente os dados do Brasil com a função loc
brasil = dados.loc['Brasil', anos]
brasil
```

```python
#cria um dicionario com a coluna ano e imigrantes
brasil_dicionario = { 'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist() }

#cria o dataframe a partir do dicionário criado anteriormente
dados_brasil = pd.DataFrame(brasil_dicionario)

dados_brasil
```

## Criando gráficos com o matplotlib

```python
import matplotlib.pyplot as plt
```

```python
#ajustando a largura e altura do gráfico
plt.figure(figsize=(10,3))

#gerando gráfico de linhas com o ano no eixo x e os imigrantes no eixo y
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])

#ajustando o intervalo dos dados no eixo x
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])

#ajustando o intervalo dos dados no eixo y
#plt.yticks([500, 1000, 1500, 2000, 2500, 3000])

#adicionando rótulos
plt.title('Imigração do Brasil para o Canadá (1980 a 2013)')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')

#exibindo o gráfico
plt.show()

```

## Comparando a imigração de Brasil e Argentina para o Canadá

```python
#criando dataframe da Argentina
dados_argentina = pd.DataFrame(dados.loc['Argentina', anos]).reset_index()
dados_argentina.columns = [['anos', 'imigrantes']]
dados_argentina
```

```python
#criando visualização para comparar Argentina e Brasil

#ajustando a largura e altura do gráfico
plt.figure(figsize=(10,5))

#gerando gráfico de linhas com o ano no eixo x e os imigrantes no eixo y
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], dados_argentina['imigrantes'])

#ajustando o intervalo dos dados no eixo x
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])

#adicionando rótulos
plt.title('Comparativo da imigração do Brasil e Argentina para o Canadá (1980 a 2013)')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.legend(['Brasil', 'Argentina'])

#exibindo o gráfico
plt.show()
```

**Há alguma tendência ou padrão comum nos dados dos dois países?**

Do final da década de 1980 para o final dos anos 2000, é possível observar que as linhas de imigração possuem um desenho muito parecido entre ambos os países.

**Quais são os períodos com maior número de imigrantes nos dois países?**

Entre os anos 2000 e 2005 a Argentina registrou os maiores números de imigrantes. Já no caso do Brasil, o maior número foi registrado no final da década de 2000.

## Analisando tendências de imigração para América do Sul

## Desafio: visualizando dados de vendas de diferentes lojas

```python
#conjunto de dados

lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
#transformando o dicionário em um dataframe
vendas_2022 = pd.DataFrame(vendas_2022)

#atribuindo as lojas como índice
vendas_2022.index = lojas

maior_venda = vendas_2022.describe()
maior_venda = max(maior_venda.loc['max'])
```

```python
#criando subplots para geração dos gráficos (matriz 2 x 2)
figura_vendas, eixo_vendas = plt.subplots(2, 2, figsize = (15, 8))
figura_vendas.suptitle('Comparando dados de diferentes lojas')
figura_vendas.subplots_adjust(hspace = 0.3)

#configurando gráfico 1
eixo_vendas[0,0].plot(vendas_2022.loc['A', vendas_2022.columns], color = 'red')
eixo_vendas[0,0].set_title('Loja A', loc = 'left', color = 'red')

#configurando gráfico 2
eixo_vendas[0,1].plot(vendas_2022.loc['B', vendas_2022.columns], color = 'blue')
eixo_vendas[0,1].set_title('Loja B', loc = 'left', color = 'blue')

#configurando gráfico 3
eixo_vendas[1,0].plot(vendas_2022.loc['C', vendas_2022.columns], color = 'green')
eixo_vendas[1,0].set_title('Loja C', loc = 'left', color = 'green')

#configurando gráfico 4
eixo_vendas[1,1].plot(vendas_2022.loc['D', vendas_2022.columns], color = 'orange')
eixo_vendas[1,1].set_title('Loja D', loc = 'left', color = 'orange')

#configurando rótulos dos eixos x e y
for eixo in eixo_vendas.flat:
  eixo.set_xlabel('Meses', fontsize = 12)
  eixo.set_ylabel('Vendas', fontsize = 12)
  eixo.spines['top'].set_visible(False)
  eixo.spines['right'].set_visible(False)

#definindo valores mínimo e máximo para cada gráfico
for eixo in eixo_vendas.ravel():
  eixo.set_ylim(0, maior_venda)

plt.show()
```

## Analisar dados da América do Sul

```python
#criando dataframe exclusivo com países da América do Sul
america_sul = dados.query('Região == "América do Sul"').sort_values('Total')

#cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

#criando uma lista de cores para destacar o Brasil em relação aos outros países
cores = []
for pais in america_sul.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')
```

```python
#criando figura
fig_america_sul, eixo_america_sul = plt.subplots(figsize = (12, 5))

#definindo os eixos do gráfico de barras
eixo_america_sul.barh(america_sul.index, america_sul['Total'], color = cores)

#definindo título
eixo_america_sul.set_title('Imigração da América do Sul para o Canadá (1980 - 2013):\nBrasil foi o 4º país com mais imigrantes', loc = 'left', fontsize = 16)

#definindo título dos rótulos
eixo_america_sul.set_xlabel('Nº de Imigrantes', fontsize = 14)
eixo_america_sul.set_ylabel('')

#ajustando tamanho dos ticks dos eixos
eixo_america_sul.xaxis.set_tick_params(labelsize = 12)
eixo_america_sul.yaxis.set_tick_params(labelsize = 12)

#adicionando anotações no gráfico
for posicao, valor in enumerate(america_sul['Total']):
  eixo_america_sul.text(valor + 20, posicao, str(valor), color = 'black', fontsize = 10, ha = 'left', va = 'center')

#retirando as bordas do gráfico
eixo_america_sul.set_frame_on(False)
#deixando invisível o eixo x
eixo_america_sul.get_xaxis().set_visible(False)
#retirando os ticks em cada barra
eixo_america_sul.tick_params(axis = 'both', which = 'both', length = 0)

fig_america_sul.savefig('imigracao_america_sul.png', transparent = False, dpi = 300, bbox_inches = 'tight')

plt.show()
```

## Visualizando os dados globais da imigração para o Canadá

```python
import seaborn as sns
```

```python
#usando tema padrão da biblioteca seaborn
sns.set_theme()
```

```python
#obtendo os top 10 que mais imigraram para o Canadá
top_10 = dados.sort_values('Total', ascending=False).head(10)
top_10
```

```python
#gerando gráfico em barras do top 10

fig_top_10, eixo_top_10 = plt.subplots(figsize = (8, 4))
eixo_top_10 = sns.barplot(data = top_10, x = 'Total', y = top_10.index, orient = 'h', palette = 'tab10')
eixo_top_10.set_title('Os 10 países com maior imigração para o Canadá\n(1980 - 2013)', loc = 'left', fontsize = 16)
eixo_top_10.set_xlabel('Número de imigrantes', fontsize = 14)
eixo_top_10.set_ylabel('')
sns.despine()
plt.show()
```

```python
#criar função para ajustar paleta de cores do gráfico
def gerar_grafico_paleta(paleta):
  fig, ax = plt.subplots(figsize = (8, 4))
  ax = sns.barplot(data = top_10, x = 'Total', y = top_10.index, orient = 'h', palette = paleta)
  eixo_top_10.set_title('Os 10 países com maior imigração para o Canadá\n(1980 - 2013)', loc = 'left', fontsize = 16)
  eixo_top_10.set_xlabel('Número de imigrantes', fontsize = 14)
  eixo_top_10.set_ylabel(' ')
  plt.show()
```

## Evolução da imigração do Brasil para o Canadá ao longo de 1980 a 2013

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes', line=dict(width=4))
)

# Definindo as configurações de layout
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir as configurações de animação
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]
fig.frames = frames

# Mostrando a figura
fig.show()
```

## Comparando a evolução ao longo do tempo de Brasil e Argentina para o Canadá

```python
#Tratando base da Argentina
df_argentina = dados.loc['Argentina'].reset_index(drop = False)
df_argentina.drop([0, 1], inplace = True)
df_argentina.reset_index(drop = True, inplace = True)
df_argentina.columns = ['ano', 'imigrantes']
df_argentina.drop([34], inplace = True)
df_argentina['ano'] = df_argentina['ano'].astype(int)
df_argentina['imigrantes'] = df_argentina['imigrantes'].astype(int)

df_argentina
```

```python
# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Brasil', line=dict(width=4)))
fig.add_trace(go.Scatter(x=[df_argentina['ano'].iloc[0]], y=[df_argentina['imigrantes'].iloc[0]], mode='lines', name='Argentina', line=dict(width=4)))

# Definindo as configurações de layout
fig.update_layout(
    title=dict(
        text='<b>Imigração de Brasil e Argentina para o Canadá (1980-2013)</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir as configurações de animação

frames = []

for i in range(len(dados_brasil)):
    frame_data = [
        go.Scatter(x = dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1]),
        go.Scatter(x = df_argentina['ano'].iloc[:i+1], y=df_argentina['imigrantes'].iloc[:i+1])
    ]
    frame = go.Frame(data = frame_data)
    frames.append(frame)

fig.frames = frames

# Mostrando a figura
fig.show()
```
